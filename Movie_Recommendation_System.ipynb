```python
import ast
```

```python
ast.literal_eval('[{"id": 28, "name": "Action"}, {"id": 12, "name": "Adventure"}, {"id": 14, "name": "Fantasy"}, {"id": 878, "name": "Science Fiction"}]')
```

```python
def convert(text):
    l=[]
    for i in ast.literal_eval(text):
        l.append(i['name'])
    return l
```

```python
movies['genres']=movies['genres'].apply(convert)
```

```python
movies['genres']
```

```python
movies.head()
```

```python
movies['keywords']=movies['keywords'].apply(convert)
```

```python
movies
```

```python
def convert_cast(text):
    l=[]
    counter=0
    for i in ast.literal_eval(text):
        if counter < 3:
            l.append(i['name'])
        counter+=1
    return l
```

```python
movies['cast']=movies['cast'].apply(convert_cast)
```

```python
movies['cast']
```

```python
movies
```

```python
def fetch_convert(text):
    l=[]
    for i in ast.literal_eval(text):
        if i['job']=='Director':
            l.append(i['name'])
            break
    return l
```

```python
movies['crew']=movies['crew'].apply(fetch_convert)
```

```python
movies['crew']
```

```python
movies.head()
```

```python
movies['overview']=movies['overview'].apply(lambda x:x.split())
```

```python
movies
```

```python
def remove_space(word):
    l=[]
    for i in word:
        l.append(i.replace(" ",""))
    return l
```

```python
movies['cast']=movies['cast'].apply(remove_space)
```

```python
movies['crew']=movies['crew'].apply(remove_space)
movies['keywords']=movies['keywords'].apply(remove_space)
movies['genres']=movies['genres'].apply(remove_space)
```

```python
movies
```

```python
movies['tags']=movies['genres']+movies['overview']+movies['cast']+movies['crew']+movies['keywords']
```

```python
new_data=movies[['movie_id','title','tags']]
```

```python
new_data
```

```python
new_data['tags']=new_data['tags'].apply(lambda x: " ".join(x))
```

```python
new_data
```

```python
movies['tags']
```

```python
new_data['tags']=new_data['tags'].apply(lambda x:x.lower())
```

```python
new_data
```

```python
new_data.shape
```

```python
import nltk
from nltk.stem import PorterStemmer
```

```python
ps=PorterStemmer()
```

```python
def stems(text):
    l=[]
    for i in text.split():
        l.append(ps.stem(i))
    return " ".join(l)
```

```python
new_data['tags']=new_data['tags'].apply(stems)
```

```python
new_data['tags']
```

```python
from sklearn.feature_extraction.text import CountVectorizer
```

```python
cv=CountVectorizer(max_features=5000,stop_words='english')
```

```python
vector=cv.fit_transform(new_data['tags']).toarray()
```

```python
vector
```

```python
vector.shape
```

```python
from sklearn.metrics.pairwise import cosine_similarity
```

```python
similarity=cosine_similarity(vector)
```

```python
similarity
```

```python
similarity.shape
```

```python
new_data[new_data['title']=='Spider-Man'].index[0]
```

```python
def recommend(movie):
    index=new_data[new_data['title']==movie].index[0]
    distances=sorted(list(enumerate(similarity[index])),reverse=True, key= lambda x:x[1])
    for i in distances[1:6]:
        print(new_data.iloc[i[0]].title)
```

```python
recommend('The Dark Knight Rises')
```
